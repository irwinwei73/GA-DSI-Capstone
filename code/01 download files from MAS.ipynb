The following downloads all Annual Reports submitted by all Singapore-registered general insurers to the Monetary Authority of Singapore.

```python
import shutil
import os
import pandas as pd
import tabula
import urllib3
import requests
from bs4 import BeautifulSoup
%config IPCompleter.greedy=True
```

```python
# read company code, prefix, name
# this csv file was manually created beforehand, information ('code' & 'company') taken fom the MAS website
# the 'prefix' value was assigned by me 
df_companies = pd.read_csv('../data/company_prefixes.csv') 
df_companies.head()
```

# Download PDF Files

```python
# Set directory path for saving files
directory_path='../data/source/' # this is fixed
try:
    os.mkdir(directory_path)
except:
    pass

# Set URL base
url_base = 'https://www.mas.gov.sg/-/media/MAS/Statistics/Insurance-Returns/annual_G/'

for index, row in df_companies.iterrows():
    for year in range(2000,2019):
        url_full = url_base + row['code'].strip() + '/' + str(year) + '/' + str(year) + '.pdf'
        print(url_full)
        filename = row['prefix'].strip() + '_' + str(year) + '.pdf'   
        print(filename)
        r = requests.get(url_full, auth=('usrname', 'password'), verify=False,stream=True)
        
        if r.status_code == 404: # file does not exist
            pass
        else:
            r.raw.decode_content = True
            with open(directory_path+'/'+filename, 'wb') as f:
                shutil.copyfileobj(r.raw, f)               
```
